```python
import numpy as np
import skfuzzy as fuzz
from skfuzzy import control as ctrl
universe = np.linspace(0,10,51)
universe1 = np.linspace(10,30,51)
food= ctrl.Antecedent(universe,'food')
service= ctrl.Antecedent(universe,'service')
tip= ctrl.Consequent(universe1,'tip')
names=['VP','P','AV','G','VG']
food.automf(names=names)
service.automf(names=names)
tip.automf(names=names)
food.view()
service.view()
tip.view()
```

```python
rule0 =ctrl.Rule(antecedent=((food['VP'] & service['VP'])|(food['VP'] & service['P'])|(food['P'] & service['VP'])),consequent=tip['VP'],label='rule P')
rule1 =ctrl.Rule(antecedent=((food['AV'] & service['VP'])|(food['G'] & service['VP'])|(food['AV'] & service['P'])|(food['P'] & service['AV'])|(food['G'] & service['VP'])),consequent=tip['P'],label='rule vp')
```

```python
from matplotlib import figure
from mpl_toolkits.mplot3d.axes3d import Axes3D as ax
import matplotlib.pyplot as plt

system = ctrl.ControlSystem([rule0,rule1])  
sim=ctrl.ControlSystemSimulation(system,flush_after_run=51*51+1)
upsampled=np.linspace(0,10,51)
x,y=np.meshgrid(upsampled,upsampled)
z=np.zeros_like(x)
for i in range(51):
  for j in range(51):
     sim.input['food']=x[i,j]
     sim.input['service']=y[i,j]
     sim.output['tip']=z[i,j]
fig = plt.figure(figsize=(8,8)) 
ax=fig.add_subplot(111, projection='3d')
suf=ax.plot_surface(x,y,z,rstride=1,cstride=1,cmap='viridis',linewidth=0.4,antialiased=True)
cset=ax.contourf(x,y,z,zdir='z',offset=9,cmap='viridis',alpha=0.5)
cset=ax.contourf(x,y,z,zdir='x',offset=11,cmap='viridis',alpha=0.5)
cset=ax.contourf(x,y,z,zdir='y',offset=11,cmap='viridis',alpha=0.5)
ax.view_init(30,200)








```
